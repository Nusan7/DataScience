## Basics of matplotlib 

It is used for visualization. Contains many plots 

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
gapminder=pd.read_csv('Data/gapminder.csv') #Importing the data. 
print(gapminder.head())
gapminder['life_exp'].astype(float) # Converting the column life_exp to float. 
life_exp=gapminder['life_exp'].values
gdp_cap=gapminder['gdp_cap'].values
print(len(life_exp))
print(len(gdp_cap))
```

```python
plt.plot(gdp_cap,life_exp)
plt.show() 
# In situations like this it is better to plot a scatter plot instead. 
```

```python
plt.scatter(gdp_cap,life_exp)
plt.xscale('log') # For a range of values logarithmic scale can be used.
plt.show()
```

### Observations: 
1. There is a correlation between the gdp_cap and life_exp. Higher the GDP, greater the life_exp 

## Histograms 
They are used to show the distribution and explore the data set. 

```python
plt.hist(life_exp,bins=15) #setting the bins increases the intervals
plt.show()
```

## Customizing the plot 

```python
print(life_exp)
print(gdp_cap)

```

```python
plt.scatter(gdp_cap,life_exp,c='r',alpha=0.8)#color customization
plt.xscale('log')
plt.xlabel('GDP') # lAbeling                                                    
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy VS GDP')
plt.yticks([0,20,40,60,80])
plt.grid()
plt.show()
```

### Boolean operators: 
1. and
2. or 
3. not

```python
a2=160
b2=1240
print(a2>a and b2<b)
print(a2==a or a2>a)
print(a2>a and not(b2<b))
```

### For numpy:
1. np.logical_or()
2. np.logical_and()
3. np.logical_not()


```python
arr1=np.array([30,40,50,60,70,80])
arr2=np.array([324,234234,50,60,34543,60])
print(np.logical_or(arr1>30,arr2<20))
```

### if,elif and else 

```python
num=2
if num>=5 and num <=10:
    print(f'The num is greater than or equal to : {num}')
elif num<5:
    print(f'The num is less than 5 as it is equal to : {num}')
else:
    print(f'The num is equal to : {num}')
```

### Loops 
1. While loop: loop until the condition is True. 
2. For loop: loop until there is no more element to loop.

```python
number=int(input('Please enter the number : \n'))
while number!=100:
    if number<100:
        print(number)
        number=number+1
    else:
        print(number)
        number=number-1
```

```python
# For loop: 
list1=['Nusan','Prapti','Dubey','Daemon','Brad','Liz']
for var in list1:
    print(var)
for index, var in enumerate(list1):
    print(f'The index is {index} and the variable is {var}')
```

```python
# For loop for dictionaries and numpy arrays:
dict1={'country':['US','China','Nepal','India'],'capitals':['Washington DC','Beijing','Kathmandu','Delhi']}
for s,t in dict1.items():
    print(s,t)
```
